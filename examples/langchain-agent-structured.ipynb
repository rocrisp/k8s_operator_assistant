```python
from langchain.llms import Ollama
llm = Ollama(model="llama3:latest", verbose=True, temperature=0)  
```

```python
from typing import Optional

from langchain.tools.base import StructuredTool

def get_last_name_for_Rany(first_name: Optional[str] = 'Rany') -> dict:

    '''Returns the last name for Rany'''

    

    info_for_Rany = {

        "first_name": "Rany",

        "last_name": "ElHousieny"

    }

    

    return info_for_Rany

last_name_for_Rany_tool = StructuredTool.from_function(get_last_name_for_Rany)        
```

```python
Rany = last_name_for_Rany_tool.run('Rany')

print(Rany)
```

```python
from langchain.agents import initialize_agent, AgentType

agent = initialize_agent(

    [last_name_for_Rany_tool],

    llm,

    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,

    verbose=True,

)
```

```python
agent.run("What is the last name for Rany?")   
```
